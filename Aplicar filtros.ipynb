```python
import pandas as pd
```

```python
base = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv', sep=';')
base
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
```

```python
df = base.query('@imoveis_comerciais not in Tipo')
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(0)
df.isnull().sum()
```

```python
apartamentos = df.query('Tipo == "Apartamento"')
itens_invalidos = apartamentos.query('Valor == 0 | Condominio == 0').index
itens_invalidos
```

```python
apartamentos.drop(itens_invalidos, axis=0, inplace=True)
apartamentos
```

```python
apartamentos.query('Valor == 0 | Condominio == 0')
```

```python
apartamentos.drop('Tipo', axis=1, inplace=True)
apartamentos
```

### 1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200


```python
apt_a = df.query('Quartos == 1 & Valor < 1200')
apt_a.to_csv('filtro1.csv', sep=';', encoding='latin1', index=False)
```

### 2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m²

```python
apt_b = df.query('Quartos > 1 & Valor < 3000 & Area > 70')
apt_b.to_csv('filtro2.csv', sep=';', encoding='latin1', index=False)
```

```python
apartamentos.to_csv('dados_apartamentos.csv', sep=';', encoding='latin1', index=False)
```
