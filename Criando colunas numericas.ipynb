```python
import pandas as pd
```

```python
base = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv', sep=';')
base
```

```python
base['Valor_por_mes'] = base['Valor'] + base['Condominio']
base
```

```python
base['Valor_por_ano'] = (base['Valor_por_mes'] * 12) + base['IPTU']
base
```

```python
base['Descricao'] = base['Tipo'] + ' em ' + base['Bairro'] + ' com ' + \
                                        base['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + base['Vagas'].astype(str) + ' vaga(s) de garagem.'
base
```

```python
base['Possui_suite'] = base['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
base
```

```python
base.to_csv('dados_completos_dev.csv', index=False, sep=';', encoding='latin1')
```
